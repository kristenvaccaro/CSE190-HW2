## Part 2: Creating actionable explanations

Imagine you work for Instagram and created this explanation for the Explore tab. Following the recommendations from prior researchers, you hoped your explanation would help build trust and transparency. 

But you know that the recommendations are bound to be inaccurate at least some of the time. And some of your test-users complained, because when they did not like the recommendation, they did not know what to do. Your goal is to provide an explanation that is **actionable**, that is, it helps users take action.

## Part 3: Gaming the system

Imagine you talk to your team about your proposal for more actionable explanations. 

One of your teammates worries that providing more actionable explanations will allow influencers to game the system. She thinks influencers will use what they learn to reverse engineer how your recommendations are made -- and then use that knowledge to get their posts artificially high in the recommendation.
